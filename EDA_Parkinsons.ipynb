# Creating a user defined function for EDA Analysis

```python
def eda_summary(dt):
  # Data Shapes

  print("The shape of the dataset is ",dt.shape)
  print("\nThe number of rows of the dataset is ",dt.shape[0])
  print("\nThe number of columns of the dataset is ",dt.shape[1])

  # First 5 rows
  print("\nThe first 5 rows of the dataset is\n",dt.head())

  # Missing values
  missing = dt.isnull().sum()
  if missing.sum() == 0:
      print("\nNo missing values found.\n")
  else:
      print(missing[missing > 0])

  # Unique Values
  print("\nNo unique values ",dt.nunique())

  # Summary of the dataset
  print("\n",dt.describe())
```

# **Dataset Overview**
The clinical dataset contains 195 records across 24 features — significantly smaller than the synthetic dataset. This small size is the primary motivation for using GANs to augment training data. No missing values were detected in the raw data, though the low unique value counts for some features (e.g., MDVP:Jitter(Abs) has only 19 unique values) suggest limited measurement resolution in those columns.

##**Feature Composition**
Features are exclusively biomedical voice measurements derived from sustained phonation recordings. They fall into three groups:
- **Frequency features**: MDVP:Fo(Hz), MDVP:Fhi(Hz), MDVP:Flo(Hz) — fundamental, maximum, and minimum vocal frequency
- **Jitter features** (frequency perturbation): MDVP:Jitter(%), MDVP:Jitter(Abs), MDVP:RAP, MDVP:PPQ, Jitter:DDP
- **Shimmer features** (amplitude perturbation): MDVP:Shimmer, MDVP:Shimmer(dB), Shimmer:APQ3, APQ5, MDVP:APQ, Shimmer:DDA
- **Nonlinear dynamics**: NHR, HNR, RPDE, DFA, spread1, spread2, D2, PPE

##**Fundamental Frequency (MDVP:Fo)**
The average vocal fundamental frequency ranges from 88.3 to 260.1 Hz with a mean of ~154.2 Hz. This is consistent with normal adult speech. The standard deviation of 41.4 Hz indicates considerable inter-subject variability, partly explained by the mix of male and female speakers in the dataset.

##**Jitter (Frequency Perturbation)**
MDVP:Jitter(%) has a mean of 0.62% and a maximum of 3.32%. Healthy individuals typically exhibit jitter below 1%, making elevated jitter a clinically meaningful indicator of vocal instability associated with Parkinson's. The right-skewed distribution (mean > median = 0.49%) suggests a subset of patients with significantly elevated jitter.

##**Shimmer (Amplitude Perturbation)**
MDVP:Shimmer has a mean of ~0.030 with a maximum of 0.119. Like jitter, elevated shimmer reflects neuromuscular control deterioration in the larynx — a hallmark of Parkinson's-related dysphonia. The distribution is also right-skewed.

##**Noise Ratios (NHR and HNR)**
NHR (Noise-to-Harmonics Ratio) has a mean of 0.025, while HNR (Harmonics-to-Noise Ratio) has a mean of 21.9 dB. Higher NHR and lower HNR both indicate increased breathiness and turbulence in voice — features strongly linked to Parkinson's vocal impairment.

##**Nonlinear Dynamics Features**
RPDE (mean 0.499) and DFA (mean 0.718) quantify the complexity and long-range correlation of vocal signals. spread1 (mean −5.68) and PPE (mean 0.207) measure fundamental frequency variation. These nonlinear features are among the most discriminative for PD detection, as Parkinson's disrupts the normal chaotic dynamics of healthy vocal production.

##**Target Variable (status)**
The mean status is ~0.754, meaning approximately 75.4% of the 195 records are Parkinson's positive (147 PD, 48 healthy) — a class imbalance of roughly 3:1. This is the core challenge the GAN addresses: the discriminator must learn to identify PD from a dataset where healthy examples are scarce.

##**Cardinality Observations**
Every continuous feature has high cardinality (148–195 unique values), consistent with real clinical measurements. The `name` column holds 195 unique patient identifiers and is dropped before modeling as it carries no predictive signal. The `status` column has exactly 2 unique values (binary classification target).

##**Key Implication**
With only 195 samples and a 3:1 class imbalance, a naive classifier could achieve ~75% accuracy by predicting Parkinson's for every sample. This makes accuracy an unreliable metric — Recall (sensitivity) is the primary metric of interest, ensuring the model minimises missed diagnoses.
